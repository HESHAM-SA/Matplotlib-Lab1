```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from datetime import datetime
import random
random.seed(0)
```

## Q1: Create two charts using list and dataframe.

```python
fruits = ['orang', 'bnana','apple']
fruits_counts = [30, 10, 50]
plt.bar(
    x = fruits ,
    height = fruits_counts,

)
plt.show();


plt.plot(
    fruits,
    fruits_counts,

)
plt.show();




price = [100,200,300]

df = pd.DataFrame({' fruits ' : fruits, 'fruits_counts': fruits_counts,'Price ':price})

df

```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(
    fruits_counts, labels=fruits,
    autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors,
    shadow=True, radius=1, explode = [0.1,0.1,0.1]
)

plt.title('Distribution of Fruits', fontsize=15)

plt.show();
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(
    fruits, fruits_counts,
     alpha=0.4, s =50
     
    
)

plt.show();
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(
    fruits, fruits_counts,
     alpha=0.4, s =50
     
    
)
plt.title("Market",color ='r', loc='left',fontsize=16)
plt.xlabel('Fruits',fontsize=16)
plt.ylabel('Fruits numbers',fontsize=16)

plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
df = pd.DataFrame(
    {
        'Products' : ['Prod 1', 'Prod 2', 'Prod 3', 'Prod 4', 'Prod 5'],
        'y1'       : [34, 56, 12, 89, 67],
        'y2'       : [12, 56, 78, 45, 90],
        'y3'       : [14, 23, 45, 25, 89]
    }
)

width = 0.2

plt.bar(np.arange(len(df))-width, df['y1'], width, color='cyan')
plt.bar(np.arange(len(df)),       df['y2'], width, color='orange')
plt.bar(np.arange(len(df))+width, df['y3'], width, color='green')

plt.xticks(np.arange(len(df)), list(df['Products'].value_counts().index))

plt.xlabel("Products")
plt.ylabel("Benifits")
plt.legend(["y1", "y2", "y3"])
plt.show();
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 10))

axs[0].bar(fruits, fruits_counts)


axs[1].pie(fruits_counts, labels=fruits)


axs[2].scatter(fruits, fruits_counts)

plt.show()
```
